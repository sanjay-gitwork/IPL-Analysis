```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ball_df = pd.read_csv("./data/IPL Ball-by-Ball 2008-2020.csv")
df = pd.read_csv("./data/IPL Matches 2008-2020.csv")
```

```python
print(df)
```

```python
df.head()
```

```python
# Dropping unnecessary columns
# df.drop(['id', 'neutral_venue','method'], axis=1, inplace=True)

# Converting date column to datetime format
# df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y')

# Handling missing values, if any
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df['id'] = df.reset_index().index
```

```python
df.head()
```

```python
df.tail()
```

```python
total_matches = df['id'].value_counts()
print(total_matches)
```

```python
# Calculate the total number of matches played by each team
matches_played = pd.concat([df['team1'], df['team2']]).value_counts()
print(matches_played)
```

```python
matches_won = df['winner'].value_counts()
print(matches_won)
```

```python
#  Percentage of matches won by each team:
matches_won_percentage = (matches_won / matches_played) * 100

# Sort the data by win percentage in descending order
matches_won_percentage = matches_won_percentage.sort_values(ascending=False)

print(matches_won_percentage)
```

```python
# Visualizing the matches won by each team:

plt.figure(figsize=(10, 6))
matches_won.plot(kind='bar')
plt.xlabel('Teams')
plt.ylabel('Number of Matches Won')
plt.title('Matches Won by Each Team')
plt.show()
```

```python
# Result distribution (wickets/runs) of matches
result_distribution = df['result'].value_counts()
print(result_distribution)
```

```python
#Visualizing the result distribution:

plt.figure(figsize=(6, 6))
result_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Result Distribution')
plt.axis('equal')
plt.show()
```

```python
# Extract the year from the date column to create the season column
df['season'] = pd.DatetimeIndex(df['date']).year
```

```python
#Total matches played each season:

matches_per_season = df['season'].value_counts().sort_index()
print(matches_per_season)
```

```python
# Determine the team that won the toss
toss_winner = df['toss_winner'].value_counts()
print(toss_winner)
```

```python
plt.figure(figsize = (10,6))
toss_winner.plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('Toss Won')
plt.title('Total no. of Toss Won by Each Team')
plt.show()
```

```python
# Calculate the percentage of matches won by each team when they won the toss
win_percentage_by_toss = (matches_won / toss_winner) * 100

# win_percentage_by_toss.sort_values(ascending=False)

# Sort the data by win percentage by toss in descending order
sorted_win_percentage_by_toss = win_percentage_by_toss.sort_values(ascending=False)


print(sorted_win_percentage_by_toss)
```

```python
plt.figure(figsize = (10,6))
sorted_win_percentage_by_toss.plot(kind = 'bar')
plt.ylabel('Toss')
plt.xlabel('Win Percentage')
plt.title('Percentage of Matches Won by Each Team By Toss')
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.hist(sorted_win_percentage_by_toss, bins=10, edgecolor='black')
plt.xlabel('Team')
plt.ylabel('Win Percentage')
plt.title('Percentage of Matches Won by Each Team')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Dropping unnecessary columns
columns_to_drop = ['id']
df2 = ball_df.drop(columns=columns_to_drop)
```

```python
df2.head()
```

```python
# Total runs scored by each team
team_runs = df2.groupby('batting_team')['total_runs'].sum()
total_team_runs = team_runs.sort_values(ascending=False)
print(total_team_runs)
```

```python
# Average runs scored per over by each team*
# team_avg_runs_per_over = df2.groupby(['batting_team', 'inning', 'over'])['total_runs'].sum().groupby(['batting_team']).mean()

# average_runs_per_over = df2.groupby(['batting_team', 'over'])['total_runs'].sum().groupby(['batting_team']).mean()


# total_team_avg_runs_per_over = average_runs_per_over.sort_values(ascending=False)
# print(average_runs_per_over)

# Calculate average runs scored by each team per over
# average_runs_per_over = df2.groupby(['batting_team', 'over'])['total_runs'].sum().mean()
# print(average_runs_per_over)


# Calculate average runs scored by each team per over
average_runs_per_over = df2.groupby(['batting_team', 'over'])['total_runs'].mean()
print(average_runs_per_over)
```

```python
# Average runs scored per innings
team_avg_runs_per_over = df2.groupby(['batting_team', 'inning'])['total_runs'].sum().groupby(['batting_team']).mean()
print(team_avg_runs_per_over)
```

```python
# Setting display options to show all rows
pd.set_option('display.max_rows', None)
```

```python
# Total runs scored by each batsman
batsman_runs = df2.groupby('batsman')['total_runs'].sum()
total_batsman_runs = batsman_runs.sort_values(ascending=False)

print(total_batsman_runs)
```

```python
# Total wickets taken by each bowler
bowler_wickets = df2[df2['is_wicket'] == 1].groupby('bowler').size()
total_bowlers_wickets = bowler_wickets.sort_values(ascending=False)
print(total_bowlers_wickets)
```

```python
# Number of balls faced by each batsman
balls_faced = df2.groupby('batsman').size()
balls_faced_each_batsman = balls_faced.sort_values(ascending=False)
print(balls_faced_each_batsman)
```

```python
# Number of over faced by each batsman
over_faced = df2.groupby('batsman').size()/6
over_faced_each_batsman = over_faced.sort_values(ascending=False)
print(over_faced_each_batsman)
```

```python
# Strike rate of each batsman
strike_rate = (total_batsman_runs / balls_faced_each_batsman) * 100
strike_rate_each_batsman = strike_rate.sort_values(ascending=False)
print(strike_rate_each_batsman)
```

```python
# Average runs scored per inning by each batsman
avg_runs_per_inning = total_batsman_runs / df2.groupby(['batsman', 'inning']).size().groupby('batsman').size()
avg_runs_per_inning_each_batsman = avg_runs_per_inning.sort_values(ascending=False)
print(avg_runs_per_inning_each_batsman)
```

```python
# Average runs conceded per over by each bowler*

avg_runs_conceded_per_over = df2.groupby(['bowler', 'inning', 'ball'])['total_runs'].sum().groupby('bowler').mean()/6
avg_runs_conceded_per_over_each_bowler = avg_runs_conceded_per_over.sort_values(ascending=False)
print(avg_runs_conceded_per_over_each_bowler)
```

```python
# Highest partnership for each team
highest_partnership = df2.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
print(highest_partnership)
```

```python
# Runs scored by each batsman against each bowler
runs_per_batsman_bowler = df2.groupby(['batsman', 'bowler'])['total_runs'].sum()
print(runs_per_batsman_bowler)
```

```python
# Number of times a batsman got dismissed
dismissals_per_batsman = df2[df2['is_wicket'] == 1].groupby('batsman').size().sort_values(ascending=False)
print(dismissals_per_batsman)
```

```python
# Average runs scored by each batsman against each bowler
avg_runs_per_batsman_bowler = runs_per_batsman_bowler / df2.groupby(['batsman', 'bowler']).size()
print(avg_runs_per_batsman_bowler)
```
